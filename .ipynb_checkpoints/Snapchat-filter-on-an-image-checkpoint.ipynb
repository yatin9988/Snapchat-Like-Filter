```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
face_classifier = cv2.CascadeClassifier("haarcascade_frontalface_alt.xml")
eye_classifier = cv2.CascadeClassifier("frontalEyes35x16.xml")
nose_classifier = cv2.CascadeClassifier("Nose18x15.xml")
```

```python
jamie = cv2.imread("Jamie_Before.jpg",-1)
```

```python
# cv2.imshow("Jamie Before",jamie)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

```python
eye = eye_classifier.detectMultiScale(jamie,1.1,5)
eye = eye[0]
```

```python
x,y,h,w = eye
```

```python
# jamie = cv2.rectangle(jamie,(x,y),(x+w,y+h),(255,255,255),2)
```

```python
# cv2.imshow("Jamie Before",jamie)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

```python
glasses = cv2.imread("glasses.png",-1)
glasses = cv2.resize(glasses,(w,h))
glass = glasses[:,:,[0,1,2]]
```

```python
plt.imshow(jamie)
plt.scatter(x,y,color="red")
plt.show()
```

```python
plt.imshow(glasses)
plt.show()
```

```python
print(glass.shape)
print(glasses.shape)
```

```python
print(w,h)
```

```python
for i in range(h):
    for j in range(w):
        if glasses[i,j,3]>0:
            jamie[y+i][x+j] = glass[i][j]
```

```python
# cv2.imshow("Jamie Before",jamie)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

```python
noses = nose_classifier.detectMultiScale(jamie,1.2,5)
print(noses)
nose = noses[0]
x,y,h,w = nose
```

```python
jamie = cv2.rectangle(jamie,(x,y),(x+w,y+h),(255,255,255),2)
```

```python
cv2.imshow("Jamie Before",jamie)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
moustache = cv2.imread("mustache.png",-1)
print(w,h)
```

```python
moustache = cv2.resize(moustache,(w,int(0.2*h)))
```

```python
plt.imshow(moustache)
plt.scatter(x,y,color="red")
plt.show()
```

```python
plt.imshow(jamie)
plt.scatter(x,y+h,color="red")
plt.show()
```

```python
for i in range(int(0.2*h)):
    for j in range(w):
        if moustache[i,j,3]>0:
            jamie[y+h+i][x+j] = moustache[i][j]
```
